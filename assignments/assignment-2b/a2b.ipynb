```python
from anomalib.data import MVTec
from anomalib.engine import Engine
from anomalib.models import EfficientAd
from anomalib.models import Patchcore
import torch

categories = ["tile", "leather", "grid"]
```

# EfficientAd

```python
for category in categories:
  # Initialize the datamodule
  datamodule = MVTec(root="/content/drive/MyDrive/f24/cs370/a2b", train_batch_size=1)
  datamodule.category = category
  datamodule.setup()

  # init/train EfficientAd model
  model = EfficientAd()
  engine = Engine(max_epochs=1)
  engine.fit(datamodule=datamodule, model=model)

  # Save checkpoint
  torch.save(engine.model.state_dict(), f"checkpoint_efficientad_{category}.ckpt")
```
